```python
import numpy as np 
import pandas as pd 
import plotly.express as px

pd.set_option('display.max_columns', 300) 
```

```python
filepath = 'data/raw/energy/'

cols = {'country':'Country', 
        'iso_code':'ISO Code', 
        'population':'Population', 
        'per_capita_electricity':'Per Capita (kWh)', 
        'primary_energy_consumption':'Energy Consumption (tWh)', 
        'renewables_share_elec':'% Renewable', 
        'year':'Year'}

df = pd.read_csv(f'data/raw/energy/World Energy Consumption.csv', usecols=cols.keys()) 

codes=['ARE', 'ARG', 'AUS', 'AUT', 'AZE', 'BEL', 'BGR', 'BLR', 'CAN', 'CHE',
       'CHL', 'COL', 'CYP', 'CZE', 'DEU', 'DNK', 'DZA', 'ECU', 'EGY', 'ESP',
       'EST', 'FIN', 'FRA', 'GBR', 'GRC', 'HKG', 'HUN', 'IRL', 'IRN', 'IRQ',
       'ISL', 'ISR', 'ITA', 'JPN', 'KAZ', 'KOR', 'KWT', 'LKA', 'LTU', 'LUX',
       'LVA', 'MAR', 'MEX', 'MYS', 'NLD', 'NOR', 'NZL', 'OMN', 'PER', 'PHL',
       'POL', 'PRT', 'QAT', 'ROU', 'SAU', 'SGP', 'SVK', 'SWE', 'THA', 'TKM',
       'TTO', 'TUR', 'TWN', 'UKR', 'UZB', 'VEN', 'VNM', 'ZAF', 
       'USA', 'RUS', 'CHN', 'IND']

df = df[(df['year'].between(1985, 2016)) 
      & (~df['country'].isin(['World', 'Asia Pacific'])) 
      & (df['iso_code'].isin(codes))
      & (df['iso_code'].notnull())
      & (df['per_capita_electricity'].notnull())]

df.rename(columns=cols, inplace=True)

df.sort_values('Year', inplace=True)

df.head()
```

```python
fig = px.scatter(df, x='Population', range_x=[0, 11e6], y='Energy Consumption (tWh)', range_y=[0, 750], 
                 color='% Renewable', size='Per Capita (kWh)', size_max=50, hover_name='Country', 
                 animation_frame='Year', animation_group='ISO Code', text='ISO Code') 

fig.update_traces(textposition='top center') 

fig.show() 
```
