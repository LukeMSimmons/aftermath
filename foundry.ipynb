```python
import numpy as np 
import pandas as pd 
import plotly.express as px

pd.set_option('display.max_columns', 300) 
```

```python
filepath = 'data/raw/cost_of_living/' 

cols = {'city': 'city', 
        'country': 'country', 
        'x1': 'Cheap Meal', 
        'x11': 'Rice (1 kg)', 
        'x23': 'Water Bottle', 
        'x33': 'Gas (1 liter)', 
        'x38': 'Internet', 
        'x48': 'Apartment (1 br)', 
        'x54': 'Avg Salary (Mo)'}

df = pd.read_csv(f'{filepath}cost-of-living.csv', usecols=cols.keys()) 
df.rename(columns=cols, inplace=True)
df.head(3) 
```

```python
countries = ['United States', 'United Kingdom', 'China', 'India', 'Russia', 'Brazil', 'Italy', 
             'Germany', 'France', 'Canada', 'Mexico', 'Japan', 'Netherlands']

df = df[(~df['Avg Salary (Mo)'].isnull()) & (df['country'].isin(countries))].copy() 
```

```python
df.sort_values(['country'], ascending=False, inplace=True)
```

```python
#all prices in (USD) 
```

```python
fig = px.scatter_3d(df, x='Cheap Meal', y='Water Bottle', z='Apartment (1 br)', size='Avg Salary (Mo)', color='country', hover_name='city', 
                    range_x=[0,50], height=600) 

camera = dict(up=dict(x=0, y=0, z=1),
              center=dict(x=-.25, y=-.25, z=0),
              eye=dict(x=-1.25, y=-1.25, z=.6))

fig.update_layout(scene_camera=camera)
fig.show() 
```

```python
df_name = 'livingcost_df' #us_health 

df.to_pickle(f'data/pkl/{df_name}.pkl')
```
