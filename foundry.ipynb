```python
import numpy as np 
import pandas as pd 
import plotly.express as px

pd.set_option('display.max_columns', 300) 
```

```python
filepath = 'data/raw/wildfires/'

df = pd.read_csv(f'{filepath}wildfires.csv') #, usecols=cols.keys()) 

df = df[(df['AcresBurned'].notnull()) & (df['Latitude'] != 0) & (df['Longitude'] < -100)]

df.head(3)
```

```python
fig = px.scatter_geo(df, lat='Latitude', lon='Longitude', size='AcresBurned', color='AcresBurned', 
                     animation_frame='ArchiveYear', hover_name='Name', size_max=50, center=dict(lat=37, lon=-121))

fig.update_geos(visible=True, scope='usa', 
    showcountries=True, countrycolor="Black",
    showsubunits=True, subunitcolor="Black", 
    showlakes=True, lakecolor="LightBlue")
fig.update_layout(margin={'r':0,'t':0,'l':0,'b':0}) 

fig.show() 
```

```python
df_name = 'wildfires_df' #us_health 

df.to_pickle(f'data/pkl/{df_name}.pkl')
```
