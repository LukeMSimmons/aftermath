```python
import numpy as np 
import pandas as pd 
import plotly.express as px

pd.set_option('display.max_columns', 300) 
```

```python
cols = {'Metacritic':'Metacritic Score', 
        'RecommendationCount':'Recommendations', 
        'ScreenshotCount':'Screenshots (Avg)', 
        'SteamSpyOwners':'Unique Owners'}
```

```python
path = 'data/raw/'

df = pd.read_csv(f'{path}steam_games.csv', usecols=cols.keys())

df.rename(columns=cols, inplace=True)

df = df[df['Recommendations'] < 4000]
df = df[df['Screenshots (Avg)'] < 41]
df = df[df['Unique Owners'] < 4e5]

df.head(3)
```

```python
dfi = df[df['Screenshots (Avg)']!=0]
px.histogram(dfi, x='Screenshots (Avg)')
```

```python
df_name = 'sfdghshf' #steam_games 

df.to_pickle(f'data/pkl/{df_name}.pkl')
```
