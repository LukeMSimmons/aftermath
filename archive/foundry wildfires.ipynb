```python
import numpy as np 
import pandas as pd 
import plotly.express as px

pd.set_option('display.max_columns', 300) 
```

```python
filepath = 'data/raw/wildfires/' 

cols = {'fire_name':'Name', 'fire_size':'Size', 'stat_cause_descr':'Cause', 
        'latitude':'Latitude', 'longitude':'Longitude', 'disc_pre_year':'Year'}

df = pd.read_csv(f'{filepath}wildfires.csv', usecols=cols.keys()) 
df.rename(columns=cols, inplace=True)

df = df[(df['Name'].notnull()) & (df['Year'] > 1999) & (~df['Cause'].isin(['Fireworks', 'Children']))].copy()

df.sort_values('Year', inplace=True)

df.head(3) 
```

```python
fig = px.scatter_geo(df, lat='Latitude', lon='Longitude', size='Size', color='Cause', 
                     animation_frame='Year', hover_name='Name', size_max=40)

fig.update_geos(visible=True, scope='usa', 
    showcountries=True, countrycolor="Black",
    showsubunits=True, subunitcolor="Black", 
    showlakes=True, lakecolor="LightBlue")

fig.update_layout(margin={'r':0,'t':0,'l':0,'b':0}) 

fig.show() 
```

```python
df_name = 'wildfires_df' #us_health 

df.to_pickle(f'data/pkl/{df_name}.pkl')
```
